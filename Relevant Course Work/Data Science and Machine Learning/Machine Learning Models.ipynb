```python
def plot_boundary(
    clf: Callable,
    X: np.ndarray,
    y: np.ndarray,
    scaler: Callable,
    title: str,
    xlabel: str = 'Age',
    ylabel: str = 'Estimated Salary',
    c: str = 'train',
    alpha: float = 0.5,
    step: float = 0.01,
    ax = None
):
    """
    Helper function for plotting the decision boundary for a trained classifier

    :param clf: `Classifier` - The trained model/estimator
    :param X: Input data (assuming normalized input data)
    :param y: The real labels for the input data
    :param scaler: The scaler used for normalizing the original training data
    :param title: The title for the plot
    :param xlabel: The label for the x-axis
    :param ylabel: The label for the y-axis
    :param c: The color mode (either `train` or `test`)
    :param alpha: Opacity for decision areas
    :param step: Discretization step for the input space
    :param ax: If passed, the axis on which the plot should be created
    """

    # If no ax is passed, then create one for a new plot
    if ax is None:
        fig, ax = plt.subplots(figsize=(6, 4.5))

        # Flag for showing the plot and its legend
        show = True
    else:
        show = False

    colors = ('red', 'green') if c == 'train' else ('tab:blue', 'tab:orange')
    cmap = ListedColormap(colors)

    # Get the input space bounds
    start_x1, start_x2 = X.min(axis=0)
    stop_x1,  stop_x2  = X.max(axis=0)

    # Create the 2D coordinates
    xx, yy = np.meshgrid(np.arange(start=start_x1 * 1.2, stop=stop_x1 * 1.2, step=step),
                         np.arange(start=start_x2 * 1.2, stop=stop_x2 * 1.2, step=step))

    # Reshape the whole input space into 2D array
    xy = np.array([xx.ravel(), yy.ravel()]).T

    # Make predictions for the entire input space
    y_pred = clf.predict(xy).reshape(xx.shape)

    # Rescale the coordinates to match the original magnitude of the data
    xx = scaler.inverse_transform(xy)[:, 0].reshape(xx.shape)
    yy = scaler.inverse_transform(xy)[:, 1].reshape(yy.shape)

    # Plot the contour of the decision boundaries
    ax.contourf(xx, yy, y_pred, alpha=alpha, cmap=cmap)

    # Un-normalize the input data
    X_scaled = scaler.inverse_transform(X)

    # Plot the samples and color them according to the correct label (not the predicted one)
    for i, j in enumerate(np.unique(y)):
        ax.scatter(X_scaled[y==j, 0], X_scaled[y==j, 1], c=colors[i], label=j,
                    s=60, edgecolors='0.2', linewidth=0.7)

    ax.axis(xmin=xx.min(), xmax=xx.max())
    ax.axis(ymin=yy.min(), ymax=yy.max())

    ax.set_title(title, fontsize=14)
    ax.set(xlabel=xlabel, ylabel=ylabel)

    if show:
        ax.legend(fontsize=13)
        plt.show()
```

```python
def plot_confusion_matrix(y_test: np.ndarray, y_pred: np.ndarray, labels: list, clf_name: str):

    # Get the confusion matrix for the current set of predictions
    cm = confusion_matrix(y_test, y_pred, labels=labels)

    # Plot the confusion matrix
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
    disp.plot(cmap='Blues')
    plt.show()

    print(f"{clf_name} accuracy: {accuracy_score(y_test, y_pred):.4f}")
```

# Decision Tree

### Teorie

* Un **arbore de decizie** este un model de nvare supervizat non-parametric, ce poate fi folosit att pentru clasificare ct i pentru regresie. Scopul este acela de a construi un model pentru prezicerea unei variabile int prin stabilirea unui set de reguli de decizie simple deduse pe baza caracteristicilor datelor de antrenare. Un arbore de decizie poate fi vzut ca o aproximare constant pe pri.


* Arborele de decizie are o structur ierarhic, iar construirea acestuia presupune o strategie de tipul divide et impera, cu o abordare lacom (greedy) n identificarea punctului optim de ramificare a arborelui. Prin divizarea recursiv a setului de date n subgrupuri pe baza caracteristicilor semnificative, arborele formeaz o structur ierarhic de decizii. Nodurile reprezint decizii, iar ramurile indic rezultate posibile. Aceast abordare ofer o interpretare intuitiv a procesului de decizie, facilitnd analiza i nelegerea modelului.


* Terminologie:
    - **Root node**: baza arborelui de decizie.
    - **Splitting**: procesul de divizare a unui nod n subgrupuri, bazat pe caracteristici specifice, pentru a continua construcia arborelui.
    - **Decision Node**: Punct n arborele decizional unde se ia o decizie, avnd ramuri care duc la alte noduri sau frunze.
    - **Leaf Node**: Un nod care nu este divizat n continuare i reprezint o predicie sau o clasificare.
    - **Prunning**: Procesul de eliminare a anumitor ramuri sau noduri ale arborelui pentru a preveni supranvarea i a mbunti generalizarea.
    - **Branch**: O conexiune ntre noduri reprezentnd calea de la un nod la altul.
    

* n construcia arborelui decizional, alegerea caracteristicii pentru mprirea unui nod este crucial. Acest proces este cunoscut sub numele de selecie de atribute i se folosete de o msur pentru a evalua eficiena mpririi unui nod (ncercnd s maximizeze omogenitatea grupurilor rezultate dup mprire). Dou astfel de msuri sunt:
    - **Information Gain**
    - **Gini Index**

# Support Vector Machine (SVM)

### Teorie

* Support Vector Machine (SVM) este un algoritm de nvare automat supervizat folosit att pentru clasificare, regresie, ct i pentru detecia valorilor extreme/aberante (outliers). Obiectivul algoritmului SVM este de a gsi un hiperplan ntr-un spaiu N-dimensional care clasific distinct punctele de date. Dimensiunea hiperplanului depinde de numrul de caracteristici. Dac numrul de caracteristici de intrare este egal cu 2, atunci hiperplanul este doar o linie. Dac numrul de caracteristici de intrare este egal cu trei, atunci hiperplanul devine un plan 2-D.


* O alegere rezonabil pentru cel mai bun hiperplan este cea care reprezint cea mai mare separare sau marj ntre cele dou clase.


* SVM este robust la valori aberante (outliers).


* Avantajele SVM:
    - Eficient n spaii cu dimensiuni ridicate.
    - Folosete un subset de puncte din setul de antrenare n funcia de decizie (numite vectori suport), fiind astfel eficient din punct de vedere al memoriei.
    - Versatil: diferite funcii Kernel pot fi specificate pentru funcia de decizie.

  * https://www.youtube.com/watch?v=_PwhiWxHK8o
  * https://towardsdatascience.com/https-medium-com-pupalerushikesh-svm-f4b42800e989

**Avantaje**
- SVM funcioneaz relativ bine atunci cnd exist un hiperplan clar de separare ntre clase.
- Eficient din punct de vedere al consumului de memorie.

**Dezavantaje**
- Performane mai slabe atunci cnd clasele int se suprapun.
- Deoarece clasificatorul cu vectori de suport funcioneaz prin plasarea punctelor de date deasupra i sub hiperplanul de clasificare, nu exist o explicaie probabilistic pentru clasificare.
- Algoritmul SVM nu este potrivit pentru seturi de date mari.

## Ensemble methods

### Teorie

* **Ensemble** nseamn combinarea mai multor modele. Astfel, o colecie de modele este folosit pentru a face predicii, mai degrab dect un model individual.


* Presupune dou familii de metode:

  * **Bagging**  implic antrenarea mai multor modele de baz pe eantioane de date create prin bootstrap (alegerea unui subset de date) i apoi agregarea prediciilor acestora pentru a obine un model final mai robust i mai stabil. Fiecare model este antrenat indepdendent, iar rezultatul final se bazeaz pe votul majoritar dup combinarea rezultatelor tuturor modelelor. De exemplu, **Random Forest**.

  * **Boosting**  o tehnic n care modele slabe (weak learners) sunt antrenate n mod secvenial pentru a corecta erorile modelelor anterioare. Fiecare model este ponderat n funcie de performana sa, astfel nct modelele ulterioare s se concentreze mai mult pe exemplele greite. Prin acest proces iterativ, se creeaz un model puternic care mbuntete performana general a ansamblului. De exemplu, **ADA BOOST, XG BOOST**.

### Teorie

<center>

  * Este un algoritm popular de amplificare (boosting). n acest algoritm, fiecare predictor corecteaz eroarea predecesorului su.     - Regularizare
    - O abordare unic pentru construcia arborelui de decizie
    - Algoritm Greedy aproximativ
    - Weighted Quantile Sketch
    - Sparsity-Aware Split Finding
    - nvare paralel

# (Opional) Kernel SVM

### Teorie


* Funcia de nucleu (kernel) este o metod utilizat pentru a lua datele de intrare i a le transforma n forma necesar pentru procesare. "Nucleu" este folosit datorit unui set de funcii matematice utilizate n SVM.


* Funcia de nucleu transform n general setul de date astfel nct o suprafa de decizie non-liniar s poat fi transformat ntr-o ecuaie liniar ntr-un numr mai mare de spaii dimensionale.


* Un nucleu ajut la formarea hiperplanului n dimensiunea superioar fr a crete complexitatea.


* Kernel-urile sunt o modalitate de a rezolva probleme neliniare cu ajutorul clasificatorilor liniari. Aceast este cunoscut c metod trucului nucleului (*kernel trick*). Funciile nucleului sunt utilizate c parametri n codurile SVM. Ele ajut la determinarea formei hiperplanului i a limitei de decizie.


* Exemple de kernel-uri:
    - **Linear kernel**
$$
K(x_1, x_2) = x_1.x_2
$$

    - **Polynomial kernel**
$$ K(x_1, x_2) = \left(\gamma x_1.x_2 + c\right)^d $$
$$ d - \text{gradul polinomului} $$
        
    - **Sigmoid kernel**
$$ K(x_1, x_2) = \tanh(\gamma x_1.x_2 + c) $$
        
    - **RBF kernel**
$$ K(x_1, x_2) = \exp(-\gamma\left\| x_1 - x_2 \right\|^2) $$

# III. nvarea nesupervizat

## 1. Clusterizare

## K-Means

### Teorie

* Algoritmul de clusterizare K-Means este o metod de nvare nesupervizat folosit pentru rezolvarea problemelor de clusterizare.


* K-means este un algoritm ce se bazeaz pe conceptul de centroid. Un centroid este un punct central ntr-un set de date sau ntr-un cluster. n K-Means, este media punctelor dintr-un cluster, utilizat pentru a defini poziia acestuia.


* n cadrul algorimului K-means calculm distana dintre fiecare punct i un centroid pentru a-l atribui unui cluster (cel mai apropiat centroid denot cluster-ul din care punctul aparine).


* Este un proces iterativ de atribuire a fiecrui punct de date ctre grupuri i, ncet, punctele de date sunt grupate pe baza unor caracteristici similare. Obiectivul este de a minimiza suma distanelor dintre punctele de date i centroidul clusterului, pentru a identifica grupul corect cruia ar trebui s aparin fiecare punct de date.

1. **Alegerea numrului de clustere**
* Primul pas este definirea numrului ${K}$ de clustere n care vom grupa datele. Spre exemplu, s alegem ${K=3}$


2. **Iniializarea centroizilor**
* Centroidul este centrul unui cluster, dar iniial, centrul exact al punctelor de date va fi necunoscut, aa c selectm puncte de date aleatorii i le definim ca centroizi pentru fiecare cluster. Vom iniializa 3 centroizi n setul de date.

<center>
<div style="background-color: white; padding: 10px">
  
  </div>
</center>

3. **Atribuirea punctelor celui mai apropiat cluster**

* Odat definii centroizii iniiali, urmtorul pas este de a asigna fiecrui punct din setul de date cluster-ul din care aparine. Un punct aparine cluster-ului $k$ dac centroidul cu indicele $k$ este cel mai apropiat (conform msurii de distan).

<center>
<div style="background-color: white; padding: 10px;">
    
</div>
</center>

<big>
$$
d(x, y) = \sqrt{\sum_{i=1}^n (x_i - y_i)^2}
$$
</big>

4. **Actualizarea centroizilor**
* n continuare, urmtoarea poziie a centroizilor este dat de media punctelor ce se afl n cluster-ul indicat de fiecare centroid.

<center>
<div style="background-color: white; padding: 10px;">
    
</div>
</center>

5. **Repetare pn la convergen**
* Paii 3 i 4 sunt repetai fie pn cnd poziiile acestora nu se mai schimb, fie pn cnd asignarea punctelor n clustere rmne identic ntre iteraii.

**Avantaje**
- Destul de simplu de implementat.

- Generalizeaz la clustere cu diferite proprieti (forme, mrimi)

**Dezavantaje**

- Sensibil la outlieri.

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/Mall_Customers.csv'); df.head()
```

```python
X = df.iloc[:, [3, 4]].values
k_values = range(2, 11)
```

```python
wcss = []
for i in k_values:
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    # kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42, n_init='auto')
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(k_values, wcss, marker='o')
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
silhouette_scores = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    # kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    cluster_labels = kmeans.fit_predict(X)
    silhouette_avg = silhouette_score(X, cluster_labels)
    silhouette_scores.append(silhouette_avg)
```

```python
plt.plot(k_values, silhouette_scores, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for Different Numbers of Clusters')
plt.show()
```

```python
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, init='k-means++', random_state=42)
# kmeans = KMeans(n_clusters=n_clusters, init='k-means++', random_state=42, n_init='auto')
y_pred = kmeans.fit_predict(X)
```

```python
plt.figure(figsize=(6, 4.5))
colors = ['red', 'green', 'blue', 'magenta', 'cyan']

for cluster_idx in range(n_clusters):
    plt.scatter(X[y_pred==cluster_idx, 0], X[y_pred==cluster_idx, 1], s=70,
                c=colors[cluster_idx], label=f"Cluster {cluster_idx}")

plt.scatter(kmeans.cluster_centers_[:, 0],
            kmeans.cluster_centers_[:, 1],
            s=200, c='black', label='Centroids', marker='x', lw=3)

plt.title('Clusters of customers', fontsize=16)
plt.xlabel('Annual Income (k$)', fontsize=14)
plt.ylabel('Spending Score (1-100)', fontsize=14)
plt.legend(fontsize=14, loc='center right', bbox_to_anchor=(1.4, 0.5))
plt.show()
```
