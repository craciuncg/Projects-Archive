```python
import pandas as pd
import numpy as np
from scipy.ndimage import gaussian_filter1d

import matplotlib.pyplot as plt
```

## Plotting helpers

```python
def plot_data_white(df, dataset_name, sigma=5):
    lossesD = gaussian_filter1d(df['Loss Discriminator'], sigma=sigma)
    lossesG = gaussian_filter1d(df['Loss Generator'], sigma=sigma)
    accuracy = gaussian_filter1d(df['Accuracy Whitebox'], sigma=sigma)
    
    plt.plot(lossesD)
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Loss Discriminator Whitebox - ' + dataset_name)
    plt.show()
    
    plt.plot(lossesG)
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Loss Generator Whitebox - ' + dataset_name)
    plt.show()
    
    plt.plot(accuracy)
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title('Accuracy Whitebox - ' + dataset_name)
    plt.show()

def plot_data_black(df, dataset_name, sigma=5):
    lossesD = gaussian_filter1d(df['Losses Discriminator'], sigma=sigma)
    lossesG = gaussian_filter1d(df['Losses Generator'], sigma=sigma)
    accuracy = gaussian_filter1d(df['Accurracy'], sigma=sigma)
    
    epochs = 500 * np.arange(len(df))
    
    plt.plot(epochs, lossesD)
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Loss Discriminator Blackbox - ' + dataset_name)
    plt.show()
    
    plt.plot(epochs, lossesG)
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Loss Generator Blackbox- ' + dataset_name)
    plt.show()
    
    plt.plot(epochs, accuracy)
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title('Accuracy Blackbox - ' + dataset_name)
    plt.show()
```

```python
def subplot_data_white(df, dataset_name, axes, sigma=5):
    lossesD = gaussian_filter1d(df['Loss Discriminator'], sigma=sigma)
    lossesG = gaussian_filter1d(df['Loss Generator'], sigma=sigma)
    accuracy = gaussian_filter1d(df['Accuracy Whitebox'], sigma=sigma)
    
    axes[0, 0].plot(lossesD)
    axes[0, 0].set_xlabel('Epoch')
    axes[0, 0].set_ylabel('Loss')
    axes[0, 0].set_title('Loss Discriminator Whitebox - ' + dataset_name)
    
    axes[0, 1].plot(lossesG)
    axes[0, 1].set_xlabel('Epoch')
    axes[0, 1].set_ylabel('Loss')
    axes[0, 1].set_title('Loss Generator Whitebox - ' + dataset_name)
    
    axes[0, 2].plot(accuracy)
    axes[0, 2].set_xlabel('Epoch')
    axes[0, 2].set_ylabel('Accuracy')
    axes[0, 2].set_title('Accuracy Whitebox - ' + dataset_name)

def subplot_data_black(df, dataset_name, axes, sigma=5):
    lossesD = gaussian_filter1d(df['Losses Discriminator'], sigma=sigma)
    lossesG = gaussian_filter1d(df['Losses Generator'], sigma=sigma)
    accuracy = gaussian_filter1d(df['Accurracy'], sigma=sigma)
    
    epochs = 500 * np.arange(len(df))
    
    axes[1, 0].plot(epochs, lossesD)
    axes[1, 0].set_xlabel('Epoch')
    axes[1, 0].set_ylabel('Loss')
    axes[1, 0].set_title('Loss Discriminator Blackbox - ' + dataset_name)
    
    axes[1, 1].plot(epochs, lossesG)
    axes[1, 1].set_xlabel('Epoch')
    axes[1, 1].set_ylabel('Loss')
    axes[1, 1].set_title('Loss Generator Blackbox- ' + dataset_name)
    
    axes[1, 2].plot(epochs, accuracy)
    axes[1, 2].set_xlabel('Epoch')
    axes[1, 2].set_ylabel('Accuracy')
    axes[1, 2].set_title('Accuracy Blackbox - ' + dataset_name)
```

## Results CIFAR10

### Data read

```python
df_cifar_white = pd.read_csv('results_train_gan_test1.csv')
df_cifar_black = pd.read_csv('results_black1.csv')
```

### Plot Whitebox Metrics

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(12, 8))

subplot_data_white(df_cifar_white, 'CIFAR10', axes)
subplot_data_black(df_cifar_black, 'CIFAR10', axes)

plt.tight_layout()
plt.show()
```

```python
plot_data_white(df_cifar_white, 'CIFAR10')
```

### Plot Blackbox Metrics

```python
plot_data_black(df_cifar_black, 'CIFAR10')
```
