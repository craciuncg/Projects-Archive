```python
import numpy as np
import matplotlib.pyplot as plt

def gaussian_distribution(x, mu, sigma_squared):
    return 1 / np.sqrt(2 * np.pi * sigma_squared) * np.exp(-0.5 * ((x - mu) ** 2) / sigma_squared)

# Set the mean and variance
mean = 0   # Replace with your desired mean (m)
variance = 1   # Replace with your desired variance (s^2)

# Generate x values
x_values = np.linspace(mean - 3 * np.sqrt(variance), mean + 3 * np.sqrt(variance), 1000)

# Calculate y values using the Gaussian distribution function
y_values = gaussian_distribution(x_values, mean, variance)

# Plot the Gaussian distribution
plt.plot(x_values, y_values, label=f'N({mean}, {variance})')
plt.title('Gaussian Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```
